```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
dependent
```

```python
#Splitting Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Model Creation Phase:
#Reference for HYPERTUNED PARAMETER - https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error', splitter='best')# we have to check with multiple Hypertuning parameters.
regressor=regressor.fit(X_train,y_train) #after executing this code. Then the MODEL will be Created here.
# Regressor is a variable to store all the inputs
#Default Parameter is "SQUARED ERROR" and the Splitter is "BEST"
```

```python
import matplotlib.pyplot as plt #This library is responsible for the graph purpose
from sklearn import tree
tree.plot_tree(regressor) #Here the regressor is a varaible we call it from Model Creation phase
plt.show()
```

```python
#Once the Model Creation Done we have to evaluate it
#Evaluation Phase
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
